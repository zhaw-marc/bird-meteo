## 2. Row counts

```python
conn = sqlite3.connect(DB_PATH)
for table in ['observations', 'checklists']:
    n = pd.read_sql(f'SELECT COUNT(*) as n FROM {table}', conn).iloc[0]['n']
    print(f'{table}: {n:,} rows')
conn.close()
```
